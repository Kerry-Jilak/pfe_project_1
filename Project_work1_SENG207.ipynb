## Cleaning up the data

```python
# checking for duplicates in each column
dataset.duplicated()
```

```python
# checking for null values
dataset.isnull()
dataset.isnull().sum()
```

```python
#viewing our dataset with null values and duplicates
dataset
```

```python
#changing the names of our variables to fix names in python case
dataset.columns = dataset.columns.str.replace(' ','_')
```

```python
#viewing our dataset now
dataset
```

```python
#we want to calculate the mean of all our columns to be used to fill in the gaps of our null places and also the duplicates.
meanFxd= dataset.fixed_acidity.mean()
meanVolA= dataset.volatile_acidity.mean()
meanCitA = dataset.citric_acid.mean()
meanResA= dataset.residual_sugar.mean()
meanCl = dataset.chlorides.mean()
meanFreeSD = dataset.free_sulfur_dioxide.mean()
meanTotSD = dataset.total_sulfur_dioxide.mean()
meanDen = dataset.density.mean()
meanpH = dataset.pH.mean()
meanSulp = dataset.sulphates.mean()
meanAl = dataset.alcohol.mean()
meanQ = dataset.quality.mean()
indexing = dataset.index
```

```python
#we want to view our means now in 4dp
print(f'The mean value of fixed_acidity  is {meanFxd:,.4f}')
print(f'The mean value of volatile_acidity is {meanVolA:,.4f}')
print(f'The mean value of citric_acid is {meanCitA:,.4f}')
print(f'The mean value of residual_sugar is {meanResA:,.4f}')
print(f'The mean value of chlorides is {meanCl:,.4f}')
print(f'The mean value of free_sulfur_dioxide is {meanFreeSD:,.4f}')
print(f'The mean value of total_sulfur_dioxide is {meanTotSD:,.4f}')
print(f'The mean value of density is {meanDen:,.4f}')
print(f'The mean value of pH is {meanpH:,.4f}')
print(f'The mean value of sulphate is {meanSulp:,.4f}')
print(f'The mean value of alcohol is {meanAl:,.4f}')
print(f'The mean value of quality is {meanQ:,.4f}')
```

```python
#declaring variable called mean_values so we can get all means in one place
mean_values=dataset.mean()
```

```python
mean_values
```

```python
#filling the null spaces and dropping all duplicates with our mean values
dataset = dataset.fillna(mean_values).drop_duplicates()

```

# CHARTS

```python
# Creating a pie chart  for our cleaned up data
colors = ['blue', 'violet', 'yellow', 'green', 'red', 'purple',
          'pink', 'black', 'gray', 'black', 'violet', 'cyan']
explode = (0.5,0.5,1,1,1,0,0,0.5,0,0.5,0,0.5)
meanValues = [meanFxd, meanVolA, meanCitA, meanResA, meanCl,
              meanFreeSD, meanTotSD, meanDen, meanpH, meanSulp, meanAl, meanQ]
meanTopics = ['FA',
              'VA',
              'CA',
              'RS',
              'C',
              'FSD',
              'TSD',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
xtitle = 'Column names'
ytitle = 'Mean Values'
legendLabels = ['fixed_acidity (FA)',
              'volatile_acidity (VA)',
              'citric_acid (CA)',
              'residual_sugar (RS)',
              'chlorides (C)',
              'free_sulfur_dioxide (FSD)',
              'total_sulfur_dioxide (TSD)',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
labels = meanTopics
sizes = meanValues
```

```python
# plotting the pie chart
plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', pctdistance=1.01, shadow=True, colors=colors, labeldistance=1.2)
plt.axis('equal')
plt.legend(title='index',labels=legendLabels)
plt.title('Mean Values for each of the columns in a pie chart')
plt.figure(figsize=(25,25))
plt.show()
```

```python
# plotting a bar graph using the means as our data

fig = plt.figure(figsize=(10, 5))

plt.bar(labels, sizes, width=0.5, color=colors)
plt.xlabel(xtitle)
plt.ylabel(ytitle)
plt.title('Comparing mean of columns in a bar chart')
plt.legend( loc='upper right', )
plt.show()
```
